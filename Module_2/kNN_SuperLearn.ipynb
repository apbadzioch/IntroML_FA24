## **Supervised Machine Learning Algorithms Introduction**

### 1. k-Nearest Neighbors

- The simplest ML algorithm in which building the model consists of storing the training dataset
- Makes predictions for new data points using the closest data points in the training dataset (nearest neighbor)
- Use odd numbers for k, to help manage decision making

```python
# must use pip install for mglearn
!pip install mglearn
# Import all modules needed
import sys
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
import IPython
import sklearn
import mglearn
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

#### setting up the DataFrame to view

```python
cancer_df = pd.DataFrame(data=cancer.data, columns=cancer.feature_names)
cancer_df['target'] = cancer.target
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=1)
```

```python
%matplotlib inline
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    cancer['data'], cancer['target'], stratify=cancer['target'], random_state=66)

training_accuracy = []
test_accuracy = []
neighbors_settings = range(1, 11)

for n_neighbors in neighbors_settings:
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X_train, y_train)
    training_accuracy.append(clf.score(X_train, y_train))
    test_accuracy.append(clf.score(X_test, y_test))

plt.plot(neighbors_settings, training_accuracy, label='training accuracy')
plt.plot(neighbors_settings, test_accuracy, label='test accuracy')
plt.ylabel("Accuracy")
plt.plot("n_neighbors")
plt.legend()    
```

```python
y_pred = clf.predict(X_test)
print("Test set score using mean: {:.2f}".format(np.mean(y_pred == y_test)))
print("Test set score using score: {:.2f}".format(clf.score(X_test, y_test)))
```
